```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.normal(0,1,1000)
```

```python
plt.plot(data)
```

```python
c0 = 1/data.size * ((data - data.mean())**2).sum()
```

```python
tau = 600
c=0
for i in range(data.size-tau):
        c += (data[i]-data.mean())*(data[i+tau]-data.mean())
```

```python
1/(data.size-tau)*c/c0
```
