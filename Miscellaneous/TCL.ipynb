```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dist = np.random.uniform(-1,1, int(1e6))
```

```python
sample_sizes = np.power(10, np.linspace(1, 3, 3)).astype(int)  # 10, 100, 1000, 10000
medie, stds = [], []
for n in sample_sizes:
    extractions = np.array([np.mean(np.random.choice(dist, size=n, replace=False)) for _ in range(1000)])
    medie.append(extractions.mean())
    stds.append(extractions.std())
    plt.hist(extractions, bins=30, density=True, alpha=0.6, label=f'size={n}')

plt.title("TCL: distribuzione delle medie vs dimensione del campione")
plt.xlabel("Media del campione")
plt.ylabel("Densit√†")
plt.legend()
plt.show()
```

```python
plt.plot(sample_sizes, medie, marker='o', label='Media', color='blue', linewidth=2)
plt.title("TCL: Media delle medie vs dimensione del campione")
plt.xlabel("Dimensione del campione")
plt.ylabel("Media")
plt.grid()
```

```python
plt.plot(sample_sizes, stds, marker='o', label='Deviazione standard', color='red', linewidth=2)
plt.plot(sample_sizes, 1/sample_sizes, marker='o', color='orange', linewidth=2)
plt.legend()
plt.title("TCL: Deviazione standard delle medie vs dimensione del campione")
plt.xlabel("Dimensione del campione")
plt.ylabel("Deviazione standard")
plt.grid()
```
