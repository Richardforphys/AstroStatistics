```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture

# Assuming 'data' is your histogram data
data = np.random.randn(1000)  # Example data, replace with your actual data

# Fit a GMM to the data
gmm = GaussianMixture(n_components=2)  # 2 components for the two Gaussians
gmm.fit(data.reshape(-1, 1))

# Generate the fitted Gaussian distribution
x = np.linspace(min(data), max(data), 1000).reshape(-1, 1)
pdf = np.exp(gmm.score_samples(x))  # Probability density function of the GMM

# Plot the histogram and the fitted GMM
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')
plt.plot(x, pdf, label='Fitted GMM', color='red')
plt.legend()
plt.show()
```
