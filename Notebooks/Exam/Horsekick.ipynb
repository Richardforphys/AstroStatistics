# Kicking Horses

Plot the resulting probability distribution. Careful with the normalization!\
How does it look like? Again try to fit it by hand.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
deaths = np.array([0,1,2,3,4])
groups = np.array([109,65,22,3,1])
```

```python
plt.plot(deaths, groups/groups.sum(), marker='o', linestyle='-')
plt.title('Number of Deaths by Group')
plt.xlabel('Number of Deaths')
plt.ylabel('Number of Groups')
plt.xticks(deaths)
plt.grid(True)
```

$P(n|\mu) = \frac{\mu^{n}e^{-\mu}}{n!}$ is the probability of having $n$ deaths with a mean number of deaths of $\mu = pN$ with $N =$ number of groups and $p$ is the probability of obsearving a death.

```python
from scipy.stats import poisson
```

```python
mean = np.sum(deaths * groups) / np.sum(groups)
```

```python
poiss = poisson(mean)
```

```python
plt.plot(deaths, poiss.pmf(deaths), marker='x', linestyle='--', color='red', label='Poisson PMF')
plt.bar(deaths, groups/groups.sum(), alpha=0.5, label='Observed Data', color='blue')
plt.title('Poisson Distribution Fit')
plt.xlabel('Number of Deaths')
plt.ylabel('Probability Mass Function')
plt.xticks(deaths)
plt.grid(True)
plt.legend()
```

```python
print(f'Mean of the distribution: {poiss.mean():.2f}')
print(f'Std of the distribution: {(poiss.std()):.2f}')
print(f'Sample mean: {mean:.2f}')
print(f'Sample std: {np.std(deaths):.2f}')
```
